# <font color='red'>Librairies importation</font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# dataset importation
df = pd.read_csv('Data/NHANES.csv')
```

```python
df.head()
```

```python
df.info()
```

# <font color='red'>Data Scrubbing</font>

```python
# Marital
df['DMDMARTL'] = df['DMDMARTL'].replace(
    {1.0:'Married', 2.0:'Widowed', 
    3.0:'Divorced', 4.0:'Separated',
    5.0:'Never married', 6.0:'Living with partner', 
    77.0:'Refused', np.nan:'Missing'})

# Gender
df['RIAGENDR'] = df['RIAGENDR'].replace({1:'Male', 2:'Female'})

# DMDEDUC2 variable : Academic level
df['DMDEDUC2'] = df['DMDEDUC2'].replace(
    {5.0:'College',  3.0:'HS/GED',  4.:'Some college/AA',  
     2.0:'9-11', np.nan:'Missing',  1.0:'<9',  
     9:"Don't know"})

# Inputation of missing values
df.fillna(df.mean(), inplace=True)

# check our new dataset
df.isna().sum();
```

# <font color='red'>Contingency Table</font>

```python
# Contingency Table between variables DMDEDUC2 and DMDMARTL
pd.crosstab(df['DMDEDUC2'], df['DMDMARTL'])
```

```python
# Contingency Table between variables DMDEDUC2 and DMDMARTL normalised on columns 
pd.crosstab(df['DMDEDUC2'], df['DMDMARTL']).apply(lambda z: z/z.sum(), axis=1)
```

```python
# Contingency Table between variables DMDEDUC2 and DMDMARTL normalised on lines
pd.crosstab(df['DMDEDUC2'], df['DMDMARTL']).apply(lambda z: z/z.sum(), axis=0)
```

```python
# Contingency Table between variables DMDEDUC2 and DMDMARTL based on gender
pd.crosstab([df['RIAGENDR'], df['DMDEDUC2']], df['DMDMARTL']).apply(lambda z: z/z.sum(), axis=0)
```

# <font color='red'>Bar Plots</font>

```python
plt.figure(figsize=(14, 8))
sns.countplot(x='DMDMARTL', hue='RIAGENDR', data=df)
```
